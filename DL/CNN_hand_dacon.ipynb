```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv('/content/drive/MyDrive/data/hand_movement_dacon/train.csv')
test = pd.read_csv('/content/drive/MyDrive/data/hand_movement_dacon/test.csv')
submission = pd.read_csv('/content/drive/MyDrive/data/hand_movement_dacon/sample_submission.csv')
```

```python
train
```

```python
X_train = train.drop(['id', 'target'], axis = 1)
X_test = test.drop('id', axis=1)

mins = X_train.min()
maxs = X_train.max()
X_train = (X_train - mins) / (maxs - mins)
X_test = (X_test - mins) / (maxs - mins)
```

```python
X_train
```

```python
X_train_re = np.array(X_train).reshape(-1, 1,8,4)
```

```python
X_train_re
```

```python
X_test_re = np.array(X_test).reshape(-1,1,8,4)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset

import random
random.seed(42)
torch.manual_seed(42)
```

```python
X_train_re_tensor = torch.tensor(X_train_re, dtype= torch.float32)
```

```python
X_train_re_tensor
```

```python
X_test_re_tensor = torch.tensor(X_test_re, dtype = torch.float32)
```

```python
y_train_tensor = torch.tensor(train['target'].to_numpy(), dtype = torch.int64)
```

```python
train_dataset = TensorDataset(X_train_re_tensor, y_train_tensor)
```

```python
train_dataloader = DataLoader(train_dataset, batch_size = 16, shuffle = True)
for batch_idx, samples in enumerate(train_dataloader):
    if batch_idx > 0:
        break
    print(samples[0].shape)
    print(samples[1])
```

```python
#https://justkode.kr/deep-learning/pytorch-cnn 
#위에 참고

class FlattenLayer(nn.Module):
  def forward(self, x):
    sizes = x.size()
    return x.view(sizes[0],-1)

class CNN(nn.Module):
  def __init__(self):
    super().__init__()
    self.layer1 = nn.Sequential(
        nn.Conv2d(in_channels=1, out_channels = 16, kernel_size =2, stride = 1),
        nn.BatchNorm2d(num_features=16),
        nn.ReLU(),
        nn.Conv2d(in_channels = 16, out_channels = 32, kernel_size = 2, stride =1),
        nn.BatchNorm2d(32),
        nn.ReLU()
    )
    
    self.fc = nn.Linear(in_features= 32 * 6 * 2, out_features=32 ) 
    #처음 dense를 해줄때 나왔던 out채널의 수 그대로 해줘야됨.
    #in_features 헷갈리면 forward에 out.shape 쳐서 확인.
    self.fc2 = nn.Linear(in_features = 32, out_features = 16)
    self.fc3 = nn.Linear(16,4)


  def forward(self, x):
    out = self.layer1(x)
    out = out.view(-1, 32 * 6 * 2)
    out = self.fc(out)
    out = self.fc2(out)
    out = self.fc3(out)
    return out
```

```python
model = CNN()
print(model)
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.005)

for epoch in range(100):
    
    running_loss = 0.0
    accuracy = 0
    for i, data in enumerate(train_dataloader, 0):
        inputs, labels = data

        optimizer.zero_grad() # 매개변수를 0으로 만듭니다. 매 학습시 초기화해줘야합니다.
        outputs = model(inputs) # 입력값을 넣어 순전파를 진행시킵니다.
        loss = criterion(outputs, labels) # 모델 출력값와 실제값을 손실함수에 대입합니다.
        loss.backward() # 손실함수에서 역전파 수행합니다.
        optimizer.step() # 옵티마이저를 사용해 매개변수를 최적화합니다.

        running_loss += loss.item()

        _, predictions = torch.max(outputs, 1)

        for label, prediction in zip(labels, predictions):
            if label == prediction:
                accuracy = accuracy + 1

    
    if epoch%10 == 0:
      print(f'{epoch + 1} 에포크 loss: {running_loss / i:.3f}')
      print(f'{epoch + 1} 에포크 정확도: {accuracy / (i * 16):.3f}')
```

```python
model.eval()

with torch.no_grad():
  outputs = model(X_test_re_tensor)
  _, pred = torch.max(outputs, 1)

pred
```

```python
submission['target'] = pred.numpy()
submission['target'].value_counts()
```
